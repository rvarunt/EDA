### EDA - Univariate Analysis & Multi Variate Analysis

```python
# Using Matplotlib to visualize Age distribution.

plt.style.use('ggplot')
bins = [10, 15, 20, 25, 30, 35, 40, 45, 50]
plt.hist(Cardiogoodfitness.Age, bins=bins, edgecolor='black')

plt.xlabel("Age")
plt.ylabel("Age Disribution")
plt.title("Age Distribution Histogram")
plt.show();
```

It is observed that the majority of customers are between 20 and 30 Age Group.

There are very less customers between age 45 and 50.

```python
# Using Seaborn to visualize sales by Product category.

sns.catplot(x = 'Product', data = Cardiogoodfitness, kind = 'count');
```

TM195 is the most sold product of count 80 and It is followed by TM498 with count 60.
TM798 is the least sold among other two products.

```python
# Using Seaborn to visualize Gender category.


sns.catplot(x = 'Gender', data = Cardiogoodfitness, kind = 'count');
```

It is observed there are more male consumers than female consumers who bought threadmill products from Cardio Good Fitness.

```python
# Using Seaborn to visualize Gender category.


sns.catplot(x = 'MaritalStatus', data = Cardiogoodfitness, kind = 'count');
```

Product consumption is more from couples than singles.

```python
# Using Matplotlib to visualize Income distribution.


plt.style.use('ggplot')
bins = [28000,38000,48000,58000,68000,78000,88000,98000,108000]
plt.hist(Cardiogoodfitness.Income, bins=bins, edgecolor='black')

plt.xlabel("Income")
plt.ylabel("Income Disribution")
plt.title("Income Distribution Histogram")
plt.show();
```

There are less sales purchase from High-income bracket
and majority of interests came from Mid-income group.

```python
### Multivariate Analysis
```

```python
# Using seaborn to visualize Product, Usage and Gender data.

sns.barplot(x=Cardiogoodfitness['Product'], y=Cardiogoodfitness['Usage'], hue = Cardiogoodfitness['Gender']);
```

TM798 has been used most among all the products.

```python
# Visualize products sold by geneder using matplotlib 

plt.figure(figsize=(10,10))

pd_gender = pd.crosstab(Cardiogoodfitness['Product'],Cardiogoodfitness['Gender'] )

pd_gender.plot(kind='bar')

plt.title("PRODUCTS SOLD BY GENDER")
```

TM798 is mostly bought by Males customers.
TM195 is equally bought by both Male and Female customers.

```python
# Using seaborn to visualize Miles, Income data.

sns.jointplot( x = 'Income', y = 'Miles', data = Cardiogoodfitness);
```

There is slight correlation between Income and miles. These can be due to high-income gorup can buy high end product that can be used for long distance.

```python
# Using seaborn to visualize Product, Miles and Fitness data.

sns.relplot(x="Product", y="Miles", data=Cardiogoodfitness, hue = "Fitness")
```

Observation & Inference :

It is clearly observed Product TM798 is used by high-fit group and also for long usage of upto 360 miles (miles coverage).
TM195 appeared to be used by beginners/moderately-fit groups and used for lower miles coverage upto 180 miles.
TM498 comes under mid-segment covers upto 220 miles with customer base similar to TM195.

TM798 can be inferred as high end product.

```python
# Using seaborn to visualize Product, Age and Income.

sns.scatterplot(x="Age", y="Income", data=Cardiogoodfitness, hue = "Product")
```

Observation & Inference :

It is observed that high-income group buys significant portion of TM798 and also can be inferred TM798 is expensive of all 3 products. TM195 and TM498 are preferred choice for moderatly and low-income group. There is also correlation that as age increases , average salary is also increases so group under age 22 will mostly opt for TM195 (evident in chart) and its price should be in afforable category.

```python
# Using Pairplot for view distribution, skewness, correlation and outliers.

sns.pairplot(Cardiogoodfitness,hue="Product")
```

Observations in scatterplot :

1). There is some positive correlation between Age and Income.
2). There is also some level positive correlation between Miles and Usage similarly Fitness and Usage.
3). There are outliers observed consistently in charts.

```python
# Using heatmap for correlation.

sns.heatmap(Cardiogoodfitness.corr(), annot=True)
```

Correlation observed between 

1). Miles and Usage.
2). Miles and Fitness.
3). Income and education.
4). Fitness and Usage.

```python
# Using relplot visualtion on Income, Age,Product and Usage

sns.relplot(x="Age", y="Income", hue="Product", size="Usage",
            sizes=(50, 450), alpha=.7,
            height=8, data=Cardiogoodfitness).set(title='INCOME BY AGE ,PRODUCTS AND USAGE');
```

Conclusion 

1). TM798 is high end model based on its usage, performance (Miles coverage) and income group of customers.

2). Income data has outliers; High income group prefers TM798 and Low income prefers TM195.

3). TM798's customers average age group is 29 and preferred among male customers.

4). Product sales are better in married group.

5). Active customers group are between 20 and 30.

Recommendation 

1).Target customer for TM798 is between 20 and 30 with income range above 60K. This product can be branded as high end segment.
